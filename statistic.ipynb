# Mean
- The mean value is the average value.
To calculate the mean, find the sum of all values, and divide the sum by the number of values: (99+86+87+88+111+86+103+87+94+78+77+85+86) / 13 = 89.77
- Use the NumPy mean() method to find the middle value.

````python
import numpy as np
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

m = np.mean(speed)
print(m)
````

# Median
- The median value is the value in the middle, after you have sorted all the values:
     77, 78, 85, 86, 86, 86, (87), 87, 88, 94, 99, 103, 111 
- Use the NumPy median() method to find the middle value.

````python
import numpy as np
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

m = np.median(speed)
print(m)
````

- If there are two numbers in the middle, divide the sum of those numbers by two.

````python
import numpy as np

speed = [99,86,87,88,86,103,87,94,78,77,85,86]
x = np.median(speed)
print(x)
````

# Mode
- The Mode value is the value that appears the most number of times:
99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86 = 86
- Use the SciPy mode() method to find the number that appears the most.

````python
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
x = stats.mode(speed)
print(x)
````

# Machine Learning - Standard Deviation

* Standard deviation tells how spread out the numbers are from the average.

-  Easy Example:

  Say you have marks:
 - 70, 71, 72, 73, 74 → all close to each other → low standard deviation

 - 40, 60, 80, 100, 120 → very far apart → high standard deviation

````python
# Low Standard Deviation

import numpy as np

speed = [86,87,88,86,87,85,86]
x = np.std(speed)
print(x)
````

````python
# High Standard Deviation

import numpy as np

speed = [32,111,138,28,59,77,97]
x = np.std(speed)
print(x)
````

# What is Variance

- Variance tells how spread out the data is —
it measures how far the numbers are from the mean (average).

- Example:
Let’s say you have marks:
- 70, 71, 72, 73, 74 → values are close → low variance
- 40, 60, 80, 100, 120 → values are far apart → high variance

# To calculate the variance you have to do as follows:

1. Find the mean:

```(32+111+138+28+59+77+97) / 7 = 77.4```

2. For each value: find the difference from the mean:
```python
 32 - 77.4 = -45.4
111 - 77.4 =  33.6
138 - 77.4 =  60.6
 28 - 77.4 = -49.4
 59 - 77.4 = -18.4
 77 - 77.4 = - 0.4
 97 - 77.4 =  19.6
```
3. For each difference: find the square value:
```python
(-45.4)2 = 2061.16
 (33.6)2 = 1128.96
 (60.6)2 = 3672.36
(-49.4)2 = 2440.36
(-18.4)2 =  338.56
(- 0.4)2 =    0.16
 (19.6)2 =  384.16
```
4. The variance is the average number of these squared differences:

(2061.16+1128.96+3672.36+2440.36+338.56+0.16+384.16) / 7 = 1432.2



````python
import numpy as np

speed = [32,111,138,28,59,77,97]

x = np.var(speed)

print(x)
````

- Variance → shows how spread out the data is (in squared form).
- Standard Deviation → shows how much data varies (in normal units).

-  Example:
- If your variance = 9
- then standard deviation = √9 = 3

- In short:
- Variance = spread²
- Standard Deviation = spread

- Standard Deviation is often represented by the symbol Sigma: σ

- Variance is often represented by the symbol Sigma Squared: σ2

# What are Percentiles?

- Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than.

- Example: Let's say we have an array that contains the ages of every person living on a street.

- ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

- What is the 75. percentile? The answer is 43, meaning that 75% of the people are 43 or younger.

````python
import numpy as np
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
x = np.percentile(ages, 75)
print(x)
````

````python
y = np.percentile(ages, 90)
print(y)
````

# Data Distribution

- Data distribution shows how data values are spread out —
in other words, how often each value or range of values appears in your dataset.
- Type 
1. Normal - Bell-shaped, most data near mean = Heights of people
2. UNiform - Equal chance for all values = Rolling a fair dice
3. skewed - Data leans to one side = Income data (few very rich)
4. Bimodal - Two peaks = Test scores of two groups

# Uniform Distribution

````python
import numpy as np
data = np.random.uniform(0.0, 0.5, 120)
print(data) # Create an array containing 120 random floats between 0 and 5
````

````python
import matplotlib.pyplot as plt
plt.hist(data, 5)
plt.show()
````

- We use the array from the example above to draw a histogram with 5 bars.

- The first bar represents how many values in the array are between 0 and 1.

- The second bar represents how many values are between 1 and 2.

````python
x = np.arange(len(data))
plt.scatter(x, data)
plt.show()
````

# Normal distribution

````python
import numpy as np
import matplotlib.pyplot as plt
dt = np.random.normal(0,5,100000)
plt.hist(dt, 100, color= 'skyblue') # 1000 refers how many bar you want 
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

normal_data = np.random.normal(loc=50, scale=10, size=1000)   
# mean = 50, std = 10

uniform_data = np.random.uniform(low=20, high=80, size=1000)  

# Right-Skewed Distribution (many small values, few large)
right_skewed = np.random.exponential(scale=10, size=1000)     

# Bimodal Distribution (two peaks)
bimodal_data = np.concatenate([
    np.random.normal(40, 5, 500), 
    np.random.normal(70, 5, 500)
])  

# Plot all four distributions
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Normal Distribution
axes[0, 0].hist(normal_data, bins=30, color='skyblue', edgecolor='black')
axes[0, 0].set_title("Normal Distribution")

# Uniform Distribution
axes[0, 1].hist(uniform_data, bins=30, color='lightgreen', edgecolor='black')
axes[0, 1].set_title("Uniform Distribution")

# Right-Skewed Distribution
axes[1, 0].hist(right_skewed, bins=30, color='salmon', edgecolor='black')
axes[1, 0].set_title("Right-Skewed Distribution")

# Bimodal Distribution
axes[1, 1].hist(bimodal_data, bins=30, color='violet', edgecolor='black')
axes[1, 1].set_title("Bimodal Distribution")

plt.tight_layout()
plt.show()
````
